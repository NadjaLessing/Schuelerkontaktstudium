# Plotten in Jupyter Notebook

Dieses Notebook dient als einfaches Beispiel zur Nutzung von Matplotlib zum plotten.

Viele Beispiele zur Nutzung von Matplotlib finden sich unter https://matplotlib.org/stable/gallery/index.html

```python
#Notwendige Bibliotheken werden geladen:
import numpy as np #Alle Methoden der numpy Bibliothek können wird jetzt über "np." erreichen
import matplotlib.pyplot as plt 
```

Nun wird ein einfacher Plot erstellt, bei dem y-Werte über x-Werte aufgetragen werden:

```python
plt.figure(figsize = [8,6]) #Bild wird erstellt, die Größe vom Bild wird festgelegt 
x_werte = [1, 2, 3, 4, 5] #"x-Werten" wird ein Array mit den x-Werten zugewiesen
y_werte = [2, 4, 1, 1, 3] #"y-Werten" wird eine Liste mit den y-Werten zugewiesen
plt.scatter(x_werte, y_werte) #Plotte die gegebenen y-Werte über x-Werte (als Punkte)
plt.show()
```

Statt die x-Werte manuell zu erstellen, kann auch die Methode "arange" genutzt werden.

Diese stammt aus der numpy Bibliothek, daher schreiben wir "np.arange(...)" um die Methode zu verwenden.

```python
x_werte = np.arange(1, 6, 1) #Dies erzeugt das Array [1, 2, 3, 4, 5] auf einfachere Weise
print(x_werte) #print(...) kann genutzt werden um verschiedene Objekte anzuzeigen
```

Jetzt wird der gleiche Plot nochmal erstellt:

```python
plt.figure(figsize = [8,6]) 
x_werte = np.arange(1, 6, 1) 
y_werte = [2, 4, 1, 1, 3] 
plt.scatter(x_werte, y_werte)
plt.show()
```

Die einzelnen Punkte können auch durch Linien verbunden werden.

Dazu nutzen wir statt "plt.scatter()" die Methode "plt.plot()".

```python
plt.figure(figsize = [8,6]) 
x_werte = np.arange(1, 6, 1) 
y_werte = [2, 4, 1, 1, 3] 
plt.plot(x_werte, y_werte)
plt.show()
```

Das können wir nun auch nutzen, um Graphen zu plotten:

Wir definieren die y-Werte als eine Funktion von den x-Werten.
In diesem Beispiel möchten wir $f(x) = sin(x)$  plotten.

```python
plt.figure(figsize = [8,6]) 
x_werte = np.arange(1, 6, 1) 
y_werte = np.sin(x_werte) #Dies ist der Teil, wo y = sin(x) definiert wird
plt.plot(x_werte, y_werte)
plt.show()
```

Sieht noch nicht wie ein Sinus aus?

Das liegt daran, dass wir nur fünf Punkte als x-Werte verwenden!

Wir erstellen also ein Array aus x-Werten mit mehr Werten:

```python
plt.figure(figsize = [8,6]) 
x_werte = np.arange(1, 6, 0.01) #Hier ist die Änderung: Schritte von 0.01 statt 1 in den Werten
y_werte = np.sin(x_werte) 
plt.plot(x_werte, y_werte)
plt.show()
```

Nun können wir eine Achsenbeschriftung und einen Titel hinzufügen:

```python
plt.figure(figsize = [6,4]) #Bildgröße geändert
x_werte = np.arange(1, 6, 0.01)
y_werte = np.sin(x_werte)
plt.title("Einfacher geht nicht!", fontsize = 14) #Titel und Schriftgröße festlegen
plt.xlabel("x-Achsenbeschriftung", fontsize = 12) #x-Achsenbeschriftung festlegen
plt.ylabel("y-Achsenbeschriftung", fontsize = 12) #y-Achsenbeschriftung festlegen
plt.plot(x_werte, y_werte, label = "Meine Funktion")
plt.legend()
plt.show()
```

Auch mathematische Zeichen in der Achsenbeschriftung sind kein Problem (beachte das r" ..." vor dem Schriftzug):

```python
plt.figure(figsize = [6,4]) 
x_werte = np.arange(1, 6, 0.01)
y_werte = np.sin(x_werte)
plt.title(r"$\Delta A \ll \sqrt{E^2 + E_0}$", fontsize = 14)
plt.xlabel(r"Frequenz in $\frac{1}{s}$", fontsize = 12)
plt.ylabel("Signal", fontsize = 12)
plt.plot(x_werte, y_werte, label = r"f(x) = sin(x)")
plt.legend()
plt.show()
```

Nun kann der Bereich zum Plotten noch angepasst werden:

```python
plt.figure(figsize = [6,4]) 
x_werte = np.arange(1, 100, 0.01) #x-Werte gehen nun bis 100
y_werte = np.sin(x_werte)
plt.title(r"$\Delta A \ll \sqrt{E^2 + E_0}$", fontsize = 14)
plt.xlabel(r"Frequenz in $\frac{1}{s}$", fontsize = 12)
plt.ylabel("Signal", fontsize = 12)
plt.xlim([20,40]) #Angezeigter Bereich in x-Richtung wird festgelegt
plt.ylim([-2,2]) #Angezeigter Bereich in y-Richtung wird festgelegt
plt.plot(x_werte, y_werte, label = r"f(x) = sin(x)")
plt.legend()
plt.show()
```

Eleganter ist es, die Funktion separat vom Plotten zu definieren.

```python
#Hier wird f(x) = x^2 + 2 definiert:

def f(x):
    return x**2 + 2

#Nun funktioniert das Plotten genau wie vorher.
#Die y-Werte sind nun durch f(x_werte) gegeben.

plt.figure(figsize = [6,4]) 
x_werte = np.arange(-10, 10, 0.01) #Der Bereich der x-Werte wurde angepasst
plt.title("Meine Funktion plotten", fontsize = 14)
plt.xlabel("x-Achsenbeschriftung", fontsize = 12)
plt.ylabel("y-Achsenbeschriftung", fontsize = 12)
plt.xlim([-10,10])
plt.plot(x_werte, f(x_werte), label = r"$x^{2} + 2$")
plt.legend()
plt.show()

import types

def imports():
    for name, val in globals().items():
        if isinstance(val, types.ModuleType):
            yield val.__name__

list(imports())
```
