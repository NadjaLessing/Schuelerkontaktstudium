# Einen Plot erstellen:

Nun wird ein einfacher Plot erstellt.

Dazu erstellen wir ein Bild, welches wir <code>fig</code> nennen, und Achsen, die wir <code>ax</code> nennen. <br>
Die Größe des Bildes kann mit <code>figsize = [x-Größe, y-Größe]</code> eingestellt werden.<br>
Probier mal, die Größe zu ändern!

```python
fig, ax, = plt.subplots(1, 1, figsize = [8,6]) 
```

# Mehrere Plots in einem Bild erstellen:

In <code>plt.subplots(Anzahl Zeilen, Anzahl Spalten, ...)</code> können wir die Anzahl an Plots variieren, <br> z.B 
(<code>plt.subplots(2, 3, figsize = [8,6])</code>).<br>

Die Anzahl an Achsen sollte der Anzahl der Plots angepasst werden, um die Plots einzeln ansteuern zu können <br>(sonst kommt es später zu Problemen, wenn wir was plotten wollen).<br>
Für ein Bild mit zwei Plots brauchen wir deshalb zwei Achsen, dir wir <code>ax1</code> und <code>ax2</code> nennen.

Du kannst hier mal probieren, noch mehr Plots hinzuzufügen,<br>  z.B mit <code>fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)</code>

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = [8,6])
```

# y-Werte als Punkte plotten: 

Jetzt plotten wir fünf y-Werte über fünf x-Werte.

Die x- und y-Werte schreiben wir jeweils in ein array.<br>
Zum Plotten der Werte rufen wir <code>ax.scatter</code> auf und übergeben als erstes Argument die x-Werte und als zweites Argument die y-Werte.

```python
fig, ax = plt.subplots(1, 1, figsize = [8,6])  #Bild und Plot wird erstellt, wie zuvor
x_werte = [1, 2, 3, 4, 5] #"x-Werten" wird ein array mit den x-Werten zugewiesen
y_werte = [2, 4, 1, 1, 3] #"y-Werten" wird ein array mit den y-Werten zugewiesen
ax.scatter(x_werte, y_werte) #Plotte die gegebenen y-Werte über x-Werte (als Punkte)
plt.show()
```

Statt die x-Werte manuell zu erstellen, können wir auch die Methode 
<code>numpy.arange</code> nutzen:

https://numpy.org/doc/stable/reference/generated/numpy.arange.html

Da die Methode aus der numpy Bibliothek kommt, schreiben wir 
<code>np.arange(...)
</code> 
um die Methode zu verwenden, da wir am Anfang 
<code>import numpy as np
</code>
gesagt haben.

```python
x_werte = np.arange(1, 6, 1) #Dies erzeugt das Array [1, 2, 3, 4, 5] auf einfachere Weise
print(x_werte) #print(...) kann genutzt werden um verschiedene Objekte anzuzeigen
```

Jetzt wird der gleiche Plot nochmal erstellt:

```python
fig, ax = plt.subplots(1, 1, figsize = [8,6])  
x_werte = np.arange(1, 6, 1) #Dies ist das gleiche Array wie zuvor
y_werte = [2, 4, 1, 1, 3] 
ax.scatter(x_werte, y_werte)
plt.show()
```

# y-Werte mit Linie verbinden:
Die einzelnen Punkte können auch durch Linien verbunden werden.

Dazu nutzen wir statt <code>ax.scatter</code>  die Methode <code>ax.plot</code>.

```python
fig, ax = plt.subplots(1, 1, figsize = [8,6])
x_werte = np.arange(1, 6, 1) 
y_werte = [2, 4, 1, 1, 3] 
ax.plot(x_werte, y_werte)
plt.show()
```

Das können wir nun auch nutzen, um Graphen zu plotten:

# Einen Graphen plotten:

Wir definieren die y-Werte als eine Funktion von den x-Werten.
In diesem Beispiel möchten wir $f(x) = sin(x)$  plotten.

```python
fig, ax = plt.subplots(1, 1, figsize = [8,6])
x_werte = np.arange(1, 6, 1) 
y_werte = np.sin(x_werte) #Dies ist der Teil, wo y = sin(x) definiert wird
ax.plot(x_werte, y_werte)
plt.show()
```

Sieht noch nicht wie ein Sinus aus?

Das liegt daran, dass wir nur fünf Punkte als x-Werte verwenden!

Wir erstellen also ein array aus x-Werten, welches mehr Werte enthält:

```python
fig, ax = plt.subplots(1, 1, figsize = [8,6]) 
x_werte = np.arange(1, 6, 0.01) #Hier ist die Änderung: Schritte von 0.01 statt 1 in den Werten
y_werte = np.sin(x_werte) 
ax.plot(x_werte, y_werte)
plt.show()
```

# Achsenbeschriftung und Titel hinzufügen:

Nun können wir eine Achsenbeschriftung und einen Titel hinzufügen. <br>
Außerdem wird über <code>label = "Meine Funktion" </code> eine Beschriftung in der Legende festgelegt und über <code>ax.legend()</code> die Legende geplottet.

```python
fig, ax = plt.subplots(1, 1, figsize = [6,4]) #Bildgröße geändert
x_werte = np.arange(1, 6, 0.01)
y_werte = np.sin(x_werte)
ax.set_title("Einfacher geht nicht!", fontsize = 14) #Titel und Schriftgröße festlegen
ax.set_xlabel("x-Achsenbeschriftung", fontsize = 12) #x-Achsenbeschriftung festlegen
ax.set_ylabel("y-Achsenbeschriftung", fontsize = 12) #y-Achsenbeschriftung festlegen
ax.plot(x_werte, y_werte, label = "Meine Funktion") #Beschriftung in Legende ist "Meine Funktion"
ax.legend()
plt.show()
```

Auch mathematische Zeichen in der Achsenbeschriftung sind kein Problem (beachte das r" ..." vor dem Schriftzug):

```python
fig, ax = plt.subplots(1, 1, figsize = [6,4])
x_werte = np.arange(1, 6, 0.01)
y_werte = np.sin(x_werte)
ax.set_title(r"$\Delta A \ll \sqrt{E^2 + E_0^2}$", fontsize = 14)
ax.set_xlabel(r"Frequenz in $\frac{1}{s}$", fontsize = 12)
ax.set_ylabel("Signal", fontsize = 12)
ax.plot(x_werte, y_werte, label = r"f(x) = sin(x)")
ax.legend()
plt.show()
```

# Bereich des Plots festlegen:

Nun kann der Bereich zum Plotten noch angepasst werden. <br>
Über <code>ax.set_xlim</code> und <code>ax.set_ylim</code> werden die Grenzen der Achsen festgelegt.

Du kannst mal versuchen, <code>ax.set_xlim([20,200])</code> anzugeben. <br>
In diesem Fall musst du auch deine x-Werte noch anpassen, da der Graph ja nur bis 100 geht.<br>
Also, <code>x_werte = np.arange(1, 200, 0.01)</code>.

```python
fig, ax = plt.subplots(1, 1, figsize = [6,4])
x_werte = np.arange(1, 100, 0.01) #x-Werte gehen nun bis 100
y_werte = np.sin(x_werte)
ax.set_title(r"$\Delta A \ll \sqrt{E^2 + E_0}$", fontsize = 14)
ax.set_xlabel(r"Frequenz in $\frac{1}{s}$", fontsize = 12)
ax.set_ylabel("Signal", fontsize = 12)
ax.set_xlim([20,40]) #Angezeigter Bereich in x-Richtung wird festgelegt
ax.set_ylim([-2,2]) #Angezeigter Bereich in y-Richtung wird festgelegt
ax.plot(x_werte, y_werte, label = r"f(x) = sin(x)")
ax.legend()
plt.show()
```

## Glückwunsch!
Du hast nun alles gelernt, was du brauchst um einen einfachen Graphen zu plotten. <br>
Probier doch mal noch andere Funktionen zu plotten!<br>
Falls du Sonderwünsche z.B für die Achsenbeschriftung hast, finden sich viele Beispiele auch im Internet.<br>
Diese Beispiele können oft einfach angepasst werden, wie hier der folgende Plot aus https://matplotlib.org/stable/gallery/subplots_axes_and_figures/secondary_axis.html#sphx-glr-gallery-subplots-axes-and-figures-secondary-axis-py

```python
#Alle Bibliotheken, die dieses Beispiel braucht:
import matplotlib.pyplot as plt
import numpy as np
import datetime
import matplotlib.dates as mdates
from matplotlib.ticker import AutoMinorLocator

fig, ax = plt.subplots(constrained_layout=True)

#Die x-Werte gehen von 0 bis 360 (Grad):
x = np.arange(0, 360, 1)    
#Hier wird wieder die Funktion definiert (np.pi ist die Zahl Pi):
y = np.sin(2 * x * np.pi / 180)
#Hier wird wieder y-Werte über x-Werte geplottet:
ax.plot(x, y)
#Die Achsenbeschriftung für die untere Achse
ax.set_xlabel('Winkel [Grad]')
ax.set_ylabel('Signal')
ax.set_title('Sinus')

#Hier werden zwei Funktionen yum Umrechnen von Grad zu Radiant und ungekehrt definiert:
def deg2rad(x):
    return x * np.pi / 180


def rad2deg(x):
    return x * 180 / np.pi

#Um eine Zweite Achse in Radiant zu erstellen werden die Funktionen zur Umrechnung genutzt:

secax = ax.secondary_xaxis('top', functions=(deg2rad, rad2deg))
secax.set_xlabel('Winkel [rad]')
plt.show()
```
